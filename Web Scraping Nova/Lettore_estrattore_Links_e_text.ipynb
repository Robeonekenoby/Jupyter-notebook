```python
import csv
import selenium
import time
import pandas as pd
from selenium import webdriver
```

```python
with open(r'C:\Users\robeo\Desktop\dataframeNova(senza405).csv', 'r') as csv_file:   # r read legge il file csv
    csv_reader = csv.DictReader(csv_file)
    
    links = []
    
    for line in csv_reader:
        links.append(line.get('Link'))
print(links)
    
```

```python
len(links)
```

```python

risultati = []

for link in links:
    PATH = "C:\webdrivers\chromedriver.exe"
    driver = webdriver.Chrome(PATH)

    
    driver.get(link)      
    
    #time.sleep(3)  # fa sonno per 3 secondi

    #cookie = driver.find_element_by_id('onetrust-accept-btn-handler').click() #serve altrimenti prende il p del biscotto

    


    titolo = driver.find_element_by_xpath('.//*[@id="mainbar"]/article/header/h1').text
    elementip = driver.find_elements_by_tag_name("p") #lista intera di tutti gli elementi p

    paragrafi = []
    
    for paragrafo in elementip:
        paragrafi.append(paragrafo.text)
    #print(paragrafi)


    testo = ' '.join(map(str, paragrafi))   # Use map() method for mapping str (for converting elements in list to string) with given iterator, the list.



    art_item = {
        
            'Titolo': titolo,
        
            'Testo': testo
       }
    risultati.append(art_item)

    driver.quit()
    
print(risultati)
    
    
   # RIMASUGLI DEL PROGRAMMA PASSATO    
    
    
    
    #titolo = driver.find_element_by_xpath('.//*[@id="mainbar"]/article/header/h1').text
    #testo = driver.find_element_by_xpath('.//*[@class="article-text"]').text
    #testo = driver.find_element_by_xpath('.//*[@id="mainbar"]/article/div/div/div[2]').text #si blocca a studenti 13% perch manca un div

    #titolo = driver.find_element_by_xpath('.//header/h1').text
    #testo = driver.find_element_by_xpath('.//*[@id="mainbar"]/article/div/div/div[2]').text
    #titolo = driver.find_element_by_xpath('.//article/header/h1').text
    #testo = driver.find_element_by_xpath('.//*[@class="article-text"]//p').text
    #art_item = {
        #'Link': link,
        #'Titolo': titolo,
        #'Categoria': categoria,
        #'Data_pubblicazione': data_pubblicazione,
        #'Timestamp dello scraping': timestamp,
        #'Testo': testo
   # }
    #resultati.append(art_item)

    
    #time.sleep(2)
    #driver.quit()
    
#print(resultati)




    
    
    
   # viene errore perch l'ho interrotto altrimenti ci volevano circa 4h e mezza

    
```
